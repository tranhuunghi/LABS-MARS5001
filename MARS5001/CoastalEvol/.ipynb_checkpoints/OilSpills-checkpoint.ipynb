# Oil Spill from Tanker Ships

```python
%pylab inline
import numpy as np
import datetime as dt
import netCDF4 as netCDF
import tracpy
import tracpy.plotting
from tracpy.tracpy_class import Tracpy
```

```python
# Start date in date time formatting
date1 = dt.datetime(2009, 4, 1, 0)
date2_1 = dt.datetime(2009, 4, 1, 10)
date2_2 = dt.datetime(2009, 1, 1, 10)
date3 = dt.datetime(2009, 4, 1, 18)
```

```python
# Location of TXLA model output file and grid, on a thredds server.
loc = 'http://barataria.tamu.edu:8080/thredds/dodsC/NcML/txla_nesting6.nc'
# Number of days to run the drifters.
ndays = 15
# Sets a smaller limit than between model outputs for when to force interpolation if hasn't already occurred.
nsteps = 5
# Controls the sampling frequency of the drifter tracks.
N = 4
# Use ff = 1 for forward in time and ff = -1 for backward in time.
ff = 1
ah = 0. # m^2/s
av = 0. # m^2/s
# Time between outputs
tseas = 24*3600 # 4 hours between outputs, in seconds 
# Time units
time_units = 'seconds since 1970-01-01'
# turbulence/diffusion flag
doturb = 0
# for 3d flag, do3d=0 makes the run 2d and do3d=1 makes the run 3d
do3d = 0
## Choose method for vertical placement of drifters
z0 = 's' 
num_layers = 30
zpar = num_layers-1 
```

```python
# simulation name, used for saving results into netcdf file
name1 = 'accident1'
name2_1 = 'accident2_1'
name2_2 = 'accident2_2'
name3 = 'accident3'
```

# Initialize TracPy class

```python
# Initialize Tracpy class
tp_acc1 = Tracpy(loc, name=name1, tseas=tseas, ndays=ndays, nsteps=nsteps, usebasemap=True,
            N=N, ff=ff, ah=ah, av=av, doturb=doturb, do3d=do3d, z0=z0, zpar=zpar, time_units=time_units)
tp_acc2_1 = Tracpy(loc, name=name2_1, tseas=tseas, ndays=ndays, nsteps=nsteps, usebasemap=True,
            N=N, ff=ff, ah=ah, av=av, doturb=doturb, do3d=do3d, z0=z0, zpar=zpar, time_units=time_units)
tp_acc2_2 = Tracpy(loc, name=name2_2, tseas=tseas, ndays=ndays, nsteps=nsteps, usebasemap=True,
            N=N, ff=ff, ah=ah, av=av, doturb=doturb, do3d=do3d, z0=z0, zpar=zpar, time_units=time_units)
tp_acc3 = Tracpy(loc, name=name3, tseas=tseas, ndays=ndays, nsteps=nsteps, usebasemap=True,
            N=N, ff=ff, ah=ah, av=av, doturb=doturb, do3d=do3d, z0=z0, zpar=zpar, time_units=time_units)
```

```python
# read in grid
tp_acc1._readgrid()
```

```python
# read in grid
tp_acc2_1._readgrid()
```

```python
# read in grid
tp_acc2_2._readgrid()
```

```python
# read in grid
tp_acc3._readgrid()
```

# Drifter initialization

```python
# Input starting locations as real space lon,lat locations
# Accident 1
lon1, lat1 = np.meshgrid(np.linspace(-92,-91.75,10), \
                            np.linspace(28.5,29,10)) 

# Eliminate points that are outside domain or in masked areas
lon1, lat1 = tracpy.tools.check_points(lon1, lat1, tp_acc1.grid)
```

```python
# Input starting locations as real space lon,lat locations
# Accident 2
lon2_1, lat2_1 = np.meshgrid(np.linspace(-97.,-96.75,10), \
                            np.linspace(27.,27.5,10)) 

# Eliminate points that are outside domain or in masked areas
lon2_1, lat2_1 = tracpy.tools.check_points(lon2_1, lat2_1, tp_acc2_1.grid)
```

```python
# Input starting locations as real space lon,lat locations
# Accident 2
lon2_2, lat2_2 = np.meshgrid(np.linspace(-97.,-96.75,10), \
                            np.linspace(27.,27.5,10)) 

# Eliminate points that are outside domain or in masked areas
lon2_2, lat2_2 = tracpy.tools.check_points(lon2_2, lat2_2, tp_acc2_2.grid)
```

```python
# Input starting locations as real space lon,lat locations
# Accident 3
lon3, lat3 = np.meshgrid(np.linspace(-89.75,-89.5,10), \
                            np.linspace(29,29.25,10)) 

# Eliminate points that are outside domain or in masked areas
lon3, lat3 = tracpy.tools.check_points(lon3, lat3, tp_acc3.grid)
```

# Run the numerical experiment

```python
# Note in timing that the grid was already read in
lon1p, lat1p, z1p, t1, T10, U1, V1 = tracpy.run.run(tp_acc1, date1, lon1, lat1)
```

```python
# Note in timing that the grid was already read in
lon21p, lat21p, z21p, t21, T210, U21, V21 = tracpy.run.run(tp_acc2_1, date2_1, lon2_1, lat2_1)
```

```python
# Note in timing that the grid was already read in
lon22p, lat22p, z22p, t22, T220, U22, V22 = tracpy.run.run(tp_acc2_2, date2_2, lon2_2, lat2_2)
```

```python
# Note in timing that the grid was already read in
lon3p, lat3p, z3p, t3, T30, U3, V3 = tracpy.run.run(tp_acc3, date3, lon3, lat3)
```
